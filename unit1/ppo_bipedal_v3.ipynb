```python
import gym

from huggingface_sb3 import load_from_hub, push_to_hub, package_to_hub
from huggingface_hub import notebook_login

from stable_baselines3 import PPO
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_vec_env
```

```python
env = make_vec_env("BipedalWalker-v3",n_envs=16)

model = PPO(
    "MlpPolicy",
    env=env,
    n_steps=1024,
    batch_size=512,
    n_epochs=4,
    gamma = 0.999,
    gae_lambda = 0.98,
    ent_coef = 0.01,
    verbose=1)
```

```python
model.learn(total_timesteps=1500000)
```

```python
model_name = "ppo_bipedal_v3_2"
model.save(model_name)
```

```python
# Use the following code to load the model from zip after saving
model = PPO.load("ppo_bipedal_v3", env=make_vec_env("BipedalWalker-v3", n_envs=16))
```

```python
# Now let's evaluate the trained model on a test env

test_env = gym.make("BipedalWalker-v3")
mean_reward, std_reward = evaluate_policy(model.policy,test_env,n_eval_episodes=10, deterministic=True)

print("Mean Reward is",mean_reward)
print("Standard Deviation of the Reward is",std_reward)
```
